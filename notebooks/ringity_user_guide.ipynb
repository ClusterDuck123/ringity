```python
import ringity as rng
```

## Create network

```python
N   = 2**10
rho = 0.05   # expected density
eta = 0.1    # 'randomness parameter'  : eta=1 results in an Erdos-Renyi graph.
kappa = 0    # 'concentration parameter': Higher values of kappa result in an increasing accumulation of hubs.  
a = 0.5      # 'local wiring parameter': Higher values of a result in a lower average clustering coefficient

G = rng.network_model(N=N, rho=rho, eta=eta, kappa=kappa, a=a)
```

##### Visualization

```python
rng.plot_nx(G)
```

## Diagram properties

### Persistence diagram

```python
dgm = rng.diagram(G)
print(dgm)
```

##### Visualization

```python
rng.plot_dgm(dgm)
```

### Persistence sequence (a.k.a. "Hole-Size Distribution")

```python
sequence = dgm.sequence
print(sequence)
```

##### Visualization

```python
# plot first 30 values of persistent sequence
rng.plot_seq(dgm, 30)
```

### Ring score 

```python
dgm.score
```
